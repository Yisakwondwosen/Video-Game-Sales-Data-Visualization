```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
Sales_Game=pd.read_csv('vgsales.csv.csv')
```

```python
Sales_Game
```

```python
platform_sales = Sales_Game.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(50)
```

```python
platform_sales.plot(kind='bar', ylabel='Global Sales (in millions)', title='Top Selling Platforms')
plt.show()
```

```python
Genre_sales = Sales_Game.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(50)
```

```python
Genre_sales.plot(kind='pie', ylabel='Global Sales (in millions)', title='Top Selling Platforms')
plt.show()
```

```python
Sales_Game['Year'] = Sales_Game['Year']
```

```python
platform_sales_by_year = Sales_Game.groupby(['Year', 'Platform'])['Global_Sales'].sum().unstack()
```

```python
platform_sales_by_year.plot(kind='line', marker='o', figsize=(10, 8))
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Trend of Platform Popularity Over the Years')
plt.show()
```

```python
publisher_sales = Sales_Game.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
```

```python
publisher_sales.sort_values().plot(kind='barh', figsize=(10, 8))
plt.xlabel('Global Sales')
plt.ylabel('Publisher')
plt.title('Top Publishers by Global Sales')
plt.show()
```

```python
sales_data = Sales_Game[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
```

```python
sales_data.plot(kind='box', figsize=(10, 8))
plt.ylabel('Sales (in millions)')
plt.title('Distribution of Sales Across Different Regions (Box Plot)')
plt.show()
```
